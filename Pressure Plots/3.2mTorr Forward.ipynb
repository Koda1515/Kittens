```python
pressure_y1 = []
b_field_x1 = []

for i in range (0, len(result1)):
    x = result1[i]
    dat = x[0].split('\t')
    b = float(dat[0])
    p = float(dat[1])
    b_field_x1.append(b)
    pressure_y1.append(p)
    
f.close()
```

```python
pressure_y2 = []
b_field_x2 = []

for i in range (0, len(result2)):
    x = result2[i]
    dat = x[0].split('\t')
    b = float(dat[0])
    p = float(dat[1])
    b_field_x2.append(b)
    pressure_y2.append(p)
    
f.close()
```

```python
pressure_y3 = []
b_field_x3 = []

for i in range (0, len(result3)):
    x = result3[i]
    dat = x[0].split('\t')
    b = float(dat[0])
    p = float(dat[1])
    b_field_x3.append(b)
    pressure_y3.append(p)
    
f.close()
```

```python
pressure_y4 = []
b_field_x4 = []

for i in range (0, len(result4)):
    x = result4[i]
    dat = x[0].split('\t')
    b = float(dat[0])
    p = float(dat[1])
    b_field_x4.append(b)
    pressure_y4.append(p)
    
f.close()
```

```python
import matplotlib
import matplotlib.pyplot as plt
from IPython.core.display import HTML
from IPython.core.pylabtools import figsize
from matplotlib.legend_handler import HandlerLine2D

figsize (7,4)
x1 = b_field_x1
y1 = pressure_y1

x2 = b_field_x2
y2 = pressure_y2

x3 = b_field_x3
y3 = pressure_y3

x4 = b_field_x4
y4 = pressure_y4

plt.xlabel('Magnetic Field (Gauss)')
plt.ylabel('Pressure (mTorr)')
plt.title('1.6kW_25sccm_3.2mTorr Forward')
line1, = plt.plot(x1,y1, color = '#5DADE2', marker= 'o', label= 'Manual')
line2, = plt.plot(x2,y2, color = '#AF7AC5', marker= 'o', label= 'Video 1')
line3, = plt.plot(x3,y3, color = '#58D68D', marker= 'o', label= 'Video 2')
line4, = plt.plot(x4,y4, color = '#EC7063', marker= 'o', label= 'Video 3')
plt.legend(handler_map={line1: HandlerLine2D(numpoints=1)})

plt.show()

f.close()

#Video 2, jump not continuous - flattens ~1090G to ~1130G then jumps again
#Manual, B-crit ~1010G
#Video 1, B-crit ~1030G
#video 3, B-crit ~1060G
```
