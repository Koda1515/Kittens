```python
%matplotlib notebook
import csv
import numpy as np
import os, sys
```

```python
pressure_y1 = []
b_field_x1 = []

for i in range (0, len(result)):
    x = result[i]
    dat = x[0].split('\t')
    b = float(dat[0])
    p = float(dat[1])
    b_field_x1.append(b)
    pressure_y1.append(p)
f.close()
#print(pressure_y)
```

```python
pressure_y2 = []
b_field_x2 = []

for i in range (0, len(result1)):
    x2 = result1[i]
    dat = x2[0].split('\t')
    b = float(dat[0])
    p = float(dat[1])
    b_field_x2.append(b)
    pressure_y2.append(p)
f.close()
#print(pressure_y2)
#print(pressure_y1)
```

```python
import matplotlib
import matplotlib.pyplot as plt
from IPython.core.display import HTML
from IPython.core.pylabtools import figsize
from matplotlib.legend_handler import HandlerLine2D

figsize(7,4)
x1 = b_field_x1
y1 = pressure_y1

x2 = b_field_x2
y2 = pressure_y2

plt.xlabel('Magnetic Field (Gauss)')
plt.ylabel('Pressure (mTorr)')
plt.title('1.6kW_25sccm_2.4mTorr Manual')
#plt.plot(x1,y1, 'b')
#plt.plot(x2,y2, "r")
line1, = plt.plot(x1,y1, 'b', marker = 'o', label='Forward')
line2, = plt.plot(x2,y2, 'r', marker = 'o', label='Reverse')
plt.legend(handler_map={line1: HandlerLine2D(numpoints=1)})

plt.show()

#print(x1)
f.close()
```
